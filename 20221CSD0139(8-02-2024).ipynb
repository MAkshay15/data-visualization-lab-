```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,5,1000)
print(x)
ax.plot(x,np.cos(x));
ax.plot(x,np.sin(x));
ax.plot(x,np.tan(x));
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,5,1000)
print(x)
ax.plot(x,np.cos(x));
ax.plot(x,np.sin(x));
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,5,1000)
print(x)
ax.plot(x,np.cos(x));
ax.plot(x,np.sin(x));
ax.plot(x,np.tan(x));
ax.set_title('Simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,5,1000)
print(x)
ax.plot(x,np.cos(x),label='cos');
ax.plot(x,np.sin(x),label='sin');
ax.set_title('Multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,5,1000)
print(x)
ax.plot(x,np.cos(x),label='cos',color='red');
ax.plot(x,np.sin(x),label='sin',color='pink');
ax.set_title('Multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,5,1000)
print(x)
ax.plot(x,np.cos(x),label='cos',linestyle='dashed');
ax.plot(x,np.sin(x),label='sin x+1',linestyle='dotted');
ax.set_title('Multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
